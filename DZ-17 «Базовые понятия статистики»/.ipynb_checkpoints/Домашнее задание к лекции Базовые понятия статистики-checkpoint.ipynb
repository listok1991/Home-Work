## Выбросы

```python
# функция нахождения выбросов
def outer(df,column, norm = 0.00001):
    q1 = df[column].quantile(0.25)
    q2 = df[column].quantile(0.75)
    iqr = q2 - q1
    lower_bound = q1 - (1.5 * iqr)
    upper_bound = q2 + (1.5 * iqr)
    outers_frame = df[(df[column] <= lower_bound) | (df[column] >= upper_bound)]
    outers_frame['deviation, %'] = outers_frame.apply(lambda row: (row[column] / norm - 1) * 100, axis = 1)
    return outers_frame.sort_values(column)
```

```python
outer(horse_data_select, 'pulse', 35)
```

```python
outer(horse_data_select, 'rectal temperature', 37.8)
```

```python
outer(horse_data_select, 'packed cell volume', 40)
```

К числовым данным, в которых могут быть потенциальные выбросы могут относится только три столбца:

- Температура: 14 выбросов. Из документации: "нормальная температура составляет 37,8 ", из выборки выпали значения ниже 36,6 и выше 39,6. Отклонения не превышают 10%. По информации из сети, температура может принимать все значения, указанные в выбросах.
- Гематокритичное число: найдено 5 выбросов, отклонение от нормы в районе 85%. Из документации "Уровень повышается, когда нарушается кровообращение или когда животное становится обезвоженным." Если обратить внимание на показатель "Время наполнения капилляров", то можно увидеть, что у 4 лошадей оно повышенное, что говорит о проблемах в кровообращении, соответственно данный выброс является оправданным.  

Дополнительно: на этапе описания выявлены значения в столбце "Время наполнения капилляров" - из документации возможные значения 1 и 2, при этом в двух строках найдено значение 3, что скорей всего является ошибкой данных.


## Пропуски

```python
for NAN in horse_data_select.columns:
    M = horse_data_select[NAN].isnull().sum() 
    N = M / len(horse_data_select[NAN])
 
    print(NAN, ' - ', M,  '(', round(N * 100, 1), '%)', 'пропусков')
```

- surgery? : 1 пропуск -  исключить
- rectal temperature: 60 пропусков, то есть 20% от общих данных, что существенно. Заполнение: Группировка лошадей по возрасту, и температуре конечностей. Заполнение пропусков медианой
- pulse: 24 пропуска, 8% от общих данных. Заполнение: группировка по возрасту и показателю "Переферический пульс". Заполнение: медианой.
- peripheral pulse: 69 пропусков, что составляет 23,1% от общих данных, что существенно. В документации сказано показатель является субъективным. Группировка близких по значению "Пульс".
- temperature of extremities: 56 пропусков, 18,7%. Заполнение значением 2, если показатель ректальной температуры превышает температуру (37,9). Остальные данные исключить.
- capillary refill time: 32 пропуска, 10,7%. Исключить
- packed cell volume: 29 пропусков, 9,7%. Исключить

```python
# surgery? 
horse_data_select = horse_data_select.dropna(subset=['surgery?'])
```

```python
# rectal temperature
horse_data_select['rectal temperature'].fillna(horse_data_select.groupby(['Age', 'temperature of extremities'])['rectal temperature'].transform('median'), inplace = True)
horse_data_select = horse_data_select.dropna(subset=['rectal temperature'])
```

```python
# pulse
horse_data_select['pulse'].fillna(horse_data_select.groupby(['Age', 'peripheral pulse'])['pulse'].transform('median'), inplace = True)
horse_data_select = horse_data_select.dropna(subset=['pulse'])
```

```python
# temperature of extremities
horse_data_select[(horse_data_select['rectal temperature' ] >= 37.9)]['temperature of extremities'].fillna('2')
```

```python
horse_data_select = horse_data_select.dropna(subset=['capillary refill time'])
horse_data_select = horse_data_select.dropna(subset=['peripheral pulse'])
horse_data_select = horse_data_select.dropna(subset=['temperature of extremities'])
horse_data_select = horse_data_select.dropna(subset=['capillary refill time'])
horse_data_select = horse_data_select.dropna(subset=['packed cell volume'])
```
