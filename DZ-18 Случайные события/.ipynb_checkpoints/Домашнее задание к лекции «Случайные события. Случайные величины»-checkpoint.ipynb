# Домашнее задание к лекции «Случайные события. Случайные величины»

Смоделировать игру против лотерейного автомата типа “777”. Игрок платит 1 руб., после чего выпадает случайное целое число, равномерно распределенное от 0 до 999. При некоторых значениях числа игрок получает выигрыш (см. справа)  

Выгодна ли игра игроку?  

Сколько в среднем приобретает или теряет игрок за одну игру?  

777: 200 руб.  
999: 100 руб.  
555: 50 руб.  
333: 15 руб.  
111: 10 руб.  
№77: 5 руб.  
№№7: 3 руб.  
№00: 2 руб.  
№№0: 1 руб.  

№– любая цифра   
Дополнительное задание повышенной сложности. Теоретически рассчитать средний выигрыш (проигрыш) и сравнить с результатами моделирования

```python
import numpy as np
```

```python
%%time 
trials = 10000000   # Количество попыток
win = 0 # сумма выйгрыша

for i in range(trials):
    number = np.random.randint(0, 1000)
    if number == 777:
        win += 200
    elif number == 999:
        win += 100
    elif number == 555:
        win += 50
    elif number == 333:
        win += 15
    elif number == 111:
        win += 10
    elif (number - 77) % 100 == 0:
        win += 5
    elif (number - 7) % 10 == 0:
        win += 3
    elif number % 100 == 0:
        win += 3
    elif number % 10 == 0:
        win += 1
        
# результаты эксперимента:

if win > trials:
    print('игра в автомат выгодна, сумма потраченых денег составляет', trials, 'руб., а сумма выйгрыша составила ', win, 'руб.')
else: print('игра в автомат не выгодна, сумма потраченых денег составляет', trials, 'руб., а сумма выйгрыша составила ', win, 'руб.')
print('средняя сумма выйгрыша составляет ', round(win/trials, 2), 'руб.', 'при средней стоимости игры в 1 руб. то есть средний пройгрыш/выйгрыш составляет', round(win/trials - 1, 2), 'руб.')
```
