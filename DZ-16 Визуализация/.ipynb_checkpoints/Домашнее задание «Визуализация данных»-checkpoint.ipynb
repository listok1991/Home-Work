```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
russia_alcohol = pd.read_csv('russia_alcohol.csv')
```

```python
russia_alcohol.info()
```

```python
russia_alcohol.describe()
```

```python
russia_alcohol.head()
```

```python
russia_alcohol['total'] = russia_alcohol.apply(lambda row: row['wine' : 'brandy'].sum(),axis=1) # добавление итогов по строчке
```

```python
russia_alcohol_year = russia_alcohol.groupby('year').mean() # средние показатели по всем регионам в разрезе годов
russia_alcohol_year
```

```python
#russia_alcohol_year.plot(title = 'Динамика продаж алкоголя на душу населения', xlabel = 'год') - почему то не работает аргумент xlabel
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 13,6
```

```python
plt.plot(russia_alcohol_year)
plt.title('Динамика продаж алкоголя на душу населения')
plt.xlabel('Год')
plt.ylabel('Продажи л/население')
plt.legend(russia_alcohol_year)
```

Вывод: (здесь и далее продажа = употребление) в период 1997 по 2007 растет употребление алкоголя в целом по стране. С 2007 по 2012 употребление остается на фикисрованном уровне, а с 2012 идет на спад. 
Наибольший вклад в общий показатель употребления вносит "Пиво", который и формирует общий тренд. Что можно связать с свойствами напитка (относительно низкое содержание спирта). Если рассматривать алкогольные напитки отдельно, то можно выявить следующее:  
- Употребление водки с 1997 по 2015 снижается.
- Употребление вина имеет небольшую тенденцию к росту 
- показатели напитков "Коньяк" и "Шампанское" практически не изменяются. 

```python
russia_alcohol_region = russia_alcohol.drop(['year','wine','vodka', 'champagne','beer','brandy'], axis = 'columns').groupby('region').mean().sort_values('total', axis = 0)
russia_alcohol_region
```

```python
russia_alcohol_region.plot(kind = 'bar') # общий обозор всех регионов
```

```python
plt.barh(russia_alcohol_region.head(10).index, russia_alcohol_region['total'].head(10)) # топ
plt.barh(russia_alcohol_region.tail(10).index, russia_alcohol_region['total'].tail(10)) # антитоп
plt.title('Общее употребление алкоголя в разрезе регионов')
plt.xlabel('Литр на человека')
```

Выводы: из полученных данных можно выделить топ и анти топ субьектов по продаже алкоголя. Наименее употребляющие алкоголь ожидаемо оказались регионы, с превалирующим мусульманским населением: Чеченская республика, Ингушетия, Дагестан, Северная Осетия и прочие. Так же в данный список попали регионы выбивающиеся из описанной ранее гипотезы, такие как: Севастополь и Крым. Что скорей всего связанно с отсутствием корректных данных.
Наибольшие продажи алкоголя на душу населения зафикисрованны в городах "Москва" и "Санкт-Петербург". Такой высокий показатель можно объяснить большим потоком туристов. То есть туристы увеличивают аболютные значения продажи алкоголя, при этом не учитываются в населении. 

```python
russia_alcohol_region_type = russia_alcohol.drop(['total', 'year'], axis = 'columns').groupby('region').mean()
# искуственно показатели переведем в эквивалент водки для более наглядного отражения на тепловой карте. 
# 1 л пива = 200 мл водки или 5:1
russia_alcohol_region_type['beer'] = russia_alcohol_region_type.apply(lambda row: row['beer']/5, axis=1) 
# 1 л вина = 300 мл водки или 3:1
# russia_alcohol_region_type['wine'] = russia_alcohol_region_type.apply(lambda row: row['wine']/3, axis=1) 
# 1 л шампанского = 300 мл водки или 3:1
# russia_alcohol_region_type['champagne'] = russia_alcohol_region_type.apply(lambda row: row['champagne']/3, axis=1) 

russia_alcohol_region_type
```

```python
sns.heatmap(russia_alcohol_region_type) # почему то некоторые индексы не вошли в построение тепловой карты
```

Выводы: 
1. Данные по "Брянской области" возможно не корректные
2. Во многих регионах употребление водки превосходит употребления пива (в эквиваленте водки).
3. Однако для некоторых регионов наиболее предпочитаемым напитком является именно пиво: (Рязанская обл., Воронежская обл., Новосибирская область)
4. Регионы в которых вино продается больше относительно других: Карелия, Примосркий край, Архангельская область

```python
russia_alcohol_region_type2 = russia_alcohol.drop(['year'], axis = 'columns').groupby('region').mean()
```

```python
russia_alcohol_region_type2.plot(kind = 'scatter', 
                                 x = 'vodka', 
                                 y = 'beer',
                                 s = russia_alcohol_region_type2['total']*2,
                                 title = 'Зависимость продаж Водки и Пива')
```

Выводы: Так как есть множество регионов со средними показателями продажи пива, у которых показатели продажи водки отличаются друг от друга, можно сделать вывод, что отсутствует зависимость между продажей Водки и Пива.
