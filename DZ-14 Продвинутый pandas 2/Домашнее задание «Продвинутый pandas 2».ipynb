### Задание 1  

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
import pandas as pd
```

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv')
ratings.head()
```

```python
userID = ratings.groupby('userId').count().filter(items = ['userId', 'rating'])
userID = userID[(userID['rating'] > 100)] # Создание таблицы из уникальных пользователей с количеством оценок свыше 100
```

```python
# создание двух таблиц с максимальными и минимальными значениями
max_timestamp = ratings.pivot_table(values = 'timestamp', index = 'userId', aggfunc = max)
max_timestamp.rename(columns = {'timestamp' : 'maxtimestamp'}, inplace = True)
min_timestamp = ratings.pivot_table(values = 'timestamp', index = 'userId', aggfunc = min)
min_timestamp.rename(columns = {'timestamp' : 'mintimestamp'}, inplace = True)
```

```python
new_table = userID.join(min_timestamp).join(max_timestamp) # объеденение таблиц
new_table
```

```python
new_table['longlife'] = new_table['maxtimestamp'] - new_table['mintimestamp'] #создание столбца с продолжительностью жизни
```

```python
mean = round(new_table['longlife'].mean()/60/60/24, 0)
print('среднея продолжительнсть жизни ', mean, ' дней')
```

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```
