## Задание 1

Напишите функцию, которая принимает на вход строку и проверяет является ли она валидным транспортным номером (1 буква, 3 цифры, 2 буквы, 2-3 цифры). Обратите внимание, что не все буквы кириллического алфавита используются в транспортных номерах.

Если номер валиден, то функция должна возвращать отдельно номер и регион.

Примеры работы программы:  

1. `car_id = 'АВ222С96'`

Результат: 
`Номер АВ222С валиден. Регион: 96`

2. `car_id = 'АБ22ВВ193'`

Результат: 
`Номер не валиден`

````python
import re
````

Используемые буквы в гос номерах
А, В, Е, К, М, Н, О, Р, С, Т, У и Х

````python
car_id = input('введите номер машины в формате "Х000ХХ00" или "Х000ХХ000" : ').upper() # Запрос на ввод номера
number_pattern = r'([АВЕКМНОРСТУХ]{1}[\d]{3}[АВЕКМНОРСТУХ]{2})([\d]{2,3})' # правильный шаблон номера
if len(re.findall(number_pattern, car_id)) !=0: #  проверка на ошибку, не смог придумать другой способ....
    car_id_list = list(re.findall(number_pattern, car_id)[0]) # создание списка из введенного номера
    if car_id == car_id_list[0]+car_id_list[1]: # проверка на соответствие условию
        print('Номер ', car_id_list[0], ' валиден. Регион', car_id_list[1])
    else: print('Номер не валиден')
else: print('Номер не валиден')
````

## Задание 4

Напишите функцию, которая будет принимать на вход список email-адресов и выводить их распределение по доменным зонам.

Пример работы программы:  

`emails = ['test@gmail.com, xyz@test.in, test@ya.ru, xyz@mail.ru, xyz@ya.ru', xyz@gmail.com]`

Результат: 
```
gmail.com: 2
test.in: 1 
ya.ru: 2 
mail.ru: 1
```

````python
emails = 'test@gmail.com, xyz@test.in, test@ya.ru, xyz@mail.ru, xyz@ya.ru, xyz@gmail.com'
domen_pattern = r'\@(\w+\.\w+)' # создаем шаблон для отбора доменов
emails_list = re.findall(domen_pattern, emails) # создаем список доменов
emails_dict = {}.fromkeys(emails_list, 0) # пустой список доменов для последюущей итерации
for elm in emails_list: 
    emails_dict[elm] += 1 
    
for key, value in emails_dict.items():
    print(key, ': ', value )
````
