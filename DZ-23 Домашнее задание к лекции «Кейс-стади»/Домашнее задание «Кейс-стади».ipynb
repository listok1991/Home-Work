Как критики относятся к спортивным играм?

Предположим, что если оценка игры выше 50 баллов, это означает, что отношение к игре хорошее
Нулевая гипотеза = оценки критиков спортивных игр выше 50.
Уровень значимости 5%

```python
vgsales_critic_sport = vgsales[vgsales['Genre'] ==  'Sports'].Critic_Count # фильтрация по жанру спорт и оценкам критиков
vgsales_critic_sport = vgsales_critic_sport.dropna()
alpha = 0.05
result = st.ttest_1samp(vgsales_critic_sport, float(50.0)) # почему не работает если поменять аргументы местами? (50.0, vgsales_critic_sport)
print(result)

if (result.statistic < 0) & (result.pvalue / 2 < alpha):
    print('Отвергаем нулевую гипотезу, среднее ниже 50, то есть критики низко оценивают спортивные игры')
else:
    print('Не отвергаем нулевую нулевую гипотезу, критики высоко оценивают спортивные игры')
    
print(vgsales_critic_sport.mean())
# странное p-value получилось
```

Критикам нравятся больше игры на PC или на PS4?

Сопоставим оценки двух выборок
В зависимости от знака статистики определим, какая платформа получает оценки выше. 
Нулевая гипотеза = оценки критиков не отличаются в зависимости от платформы
Уровень значимости 5%

```python
vgsales_PC = vgsales[vgsales['Platform'] ==  'PC'].Critic_Count.dropna()
vgsales_PS4 = vgsales[vgsales['Platform'] ==  'PS4'].Critic_Count.dropna()
alpha = 0.05
result = st.ttest_ind(vgsales_PC, vgsales_PS4, equal_var=False)
print(result)
if result.pvalue > alpha:
    print('не отвергаем нулевую гипотезу, оценка критиков скорей всего не зависит от платформы')
elif result.statistic < 0:
    print('отвергаем нулевую гипотезу и игры на PS3 оцениваются критиками выше')
else: 
    print('отвергаем нулевую гипотезу и игры на PC оцениваются критиками выше')
```

Критикам больше нравятся стрелялки или стратегии?

Сопоставим оценки двух выборок
В зависимости от знака статистики определим, какой жанр получает оценки выше. 
Нулевая гипотеза = оценки критиков не отличаются в зависимости от жанра
Уровень значимости 5%

```python
vgsales_Shooter = vgsales[vgsales['Genre'] ==  'Shooter'].Critic_Count.dropna()
vgsales_Strategy = vgsales[vgsales['Genre'] ==  'Strategy'].Critic_Count.dropna()
alpha = 0.05
result = st.ttest_ind(vgsales_Shooter, vgsales_Strategy, equal_var=False)
print(result)
if result.pvalue > alpha:
    print('не отвергаем нулевую гипотезу, оценка критиков скорей всего не зависит от жанра')
elif result.statistic < 0:
    print('отвергаем нулевую гипотезу и стратегии оцениваются критиками выше')
else: 
    print('отвергаем нулевую гипотезу и стрелялки оцениваются критиками выше')
```

```python
mail = pd.read_csv('spam.csv')
mail
```

```python
mail['Message'] = mail['Message'].str.lower() # приводим все символы в нижний регистр
mail
```

```python
import re
def resub(x):
    y = re.sub('[\W_]+', ' ' , x)
    return y

mail['Message'] = mail['Message'].apply(resub) # убираем все кроме букв и цифр
mail
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stopwords_set = set(stopwords.words('english'))
```

```python
# Функция удаления стоп слов
def find_stop(x):
    text = []
    y = x.split(' ')
#     print(y)
    for word in y:
        if word not in stopwords_set:
            text.append(word)
    myString = ' '.join(text)
    return myString
```

```python
mail['Message'] = mail['Message'].apply(find_stop) 
```

```python
mail
```

```python
import nltk
nltk.download('wordnet')
nltk.download('punkt')

from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
```

```python
# функция лемматизации слов
def lemmatizer(x):
    y = x.split(' ')
    text = []
    for i in y:
        word = wordnet_lemmatizer.lemmatize(i)
        text.append(word)
    myString = ' '.join(text)
    return myString
```

```python
mail['Message'] = mail['Message'].apply(lemmatizer) # Лемматизируем слова
```
